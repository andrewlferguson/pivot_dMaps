```python
import numpy as np

from scipy.spatial.distance import cdist
from scipy.linalg import eigh
from scipy.stats import gaussian_kde

import matplotlib.pyplot as plt

from sklearn.datasets import make_swiss_roll
from sklearn.neighbors import KDTree

import h5py
```

```python
class PivotDiffusionMap:
    def __init__(self, n_components, epsilon, alpha, r_cut=None, pivots=None):
        self.n_components = n_components
        self.epsilon2 = epsilon**2
        self.alpha = alpha
        if r_cut is None:
            self.r_cut2 = ( 0.9*np.sqrt(self.epsilon2) )**2
        else:        
            self.r_cut2 = r_cut**2
        if pivots is None:
            self.pivots = None
        else:
            self.pivots = pivots

    # Pivot selection
    def _select_pivots_kdtree(self, X): # NlogN scaling
        r_cut = np.sqrt(self.r_cut2)
        pivots = [X[0]]
        tree = KDTree(np.array(pivots))
        for i in range(1, X.shape[0]):
            x = X[i].reshape(1, -1)
            if tree.query_radius(x, r=r_cut, count_only=True)[0] == 0:
                pivots.append(x.flatten())
                tree = KDTree(np.array(pivots))
            if (i % 1E3 == 0):
                print('    %d points processed, %d pivot points selected' % (i, len(pivots)))
        print('    %d points processed, %d pivot points selected' % (X.shape[0], len(pivots)))
        return np.array(pivots)

    def _stream_hdf5_batches(self, file_path, group_path, batch_size): # utility function to load batches from hdf5 and place data into array for each batch
    
        with h5py.File(file_path, 'r') as f:
            group = f[group_path]
            batch_data = []
            batch_names = []
    
            # Lazy load and bactching
            for name in group:
                data = group[name][()]
                batch_data.append(data)
                batch_names.append(name)
    
                if len(batch_data) == batch_size:
                    yield batch_names, np.stack(batch_data) # stacking data into numpy array, assumes fixed length vectors
                    batch_data = []
                    batch_names = []
    
            # Handle any remaining data
            if batch_data:
                yield batch_names, np.stack(batch_data)

    def _select_pivots_kdtree_streaming(self, file_path, group_path, batch_size): # batch streaming data from hdf5 file        
        r_cut = np.sqrt(self.r_cut2)
        pivots = []

        with h5py.File(file_path, 'r') as f:
            group = f[group_path]
            num_records = len(group)
        
        ctr=0
        for names, batch_array in self._stream_hdf5_batches(file_path, group_path, batch_size):
            
            #print(f"  Loaded batch of {len(names)} samples, shape = {batch_array.shape}")
            
            for i in range(batch_array.shape[0]):
                x = batch_array[i].reshape(1, -1)
                if len(pivots) == 0:
                    pivots.append(x.flatten()) 
                    tree = KDTree(np.array(pivots))
                else:
                    if tree.query_radius(x, r=r_cut, count_only=True)[0] == 0:
                        pivots.append(x.flatten())
                        tree = KDTree(np.array(pivots))
                ctr+=1
                if (ctr % 1E3 == 0):
                    print('    %d of %d points processed, %d pivot points selected' % (ctr, num_records, len(pivots)))

        print('    %d of %d points processed, %d pivot points selected' % (ctr, num_records, len(pivots)))
        return np.array(pivots) 

    # Fit functions
    def fit(self, X):
        self.X = X

        if self.pivots is None:
            print('Commencing pivot point selection...')
            self.pivots = self._select_pivots_kdtree(X)
            print('Done!')

        self._perform_fit()

        return self

    def fit_streaming(self, file_path, group_path, batch_size):
        
        if self.pivots is None:
            print('Commencing pivot point selection...')
            self.pivots = self._select_pivots_kdtree_streaming(file_path, group_path, batch_size)
            print('Done!')

        self._perform_fit()

        return self
        
    def _perform_fit(self):

        # Compute doubly normalized pivot kernel matrix
        print('Computing pivot point diffusion map...')
        dist_PP = cdist(self.pivots, self.pivots, 'sqeuclidean')
        K_PP = np.exp(-dist_PP / self.epsilon2)
        
        r_pivots = np.sum(K_PP, axis=1)
        r_pivots_alpha = r_pivots ** self.alpha
        R_pivots_alpha_inv = np.diag(1.0 / r_pivots_alpha)
        
        c_pivots = np.sum(K_PP, axis=0)
        c_pivots_alpha = c_pivots ** self.alpha
        C_pivots_alpha_inv = np.diag(1.0 / c_pivots_alpha)
        
        K_PP_alpha = R_pivots_alpha_inv @ K_PP @ C_pivots_alpha_inv

        # Setting self.C_pivots_alpha_inv for use in Nystrom extension
        self.C_pivots_alpha_inv = C_pivots_alpha_inv

        # Confirming K_PP_alpha symmetric
        #assert np.allclose(K_PP_alpha, K_PP_alpha.T)
        #print(K_PP_alpha - K_PP_alpha.T)
        
        # Transition matrix A = D^{-1} K_PP_alpha, where D are row sums
        # Symmetric adjoint matrix A_s = D^{1/2} A D^{-1/2} = D^{-1/2} K_PP_alpha D^{-1/2}
        #  since K_PP_alpha is symmetric for all alpha
        # More stable to diagonalize As then recover evecs of A, evals are shared
        d = np.sum(K_PP_alpha, axis=1)
        D_inv_sqrt = np.diag(1.0 / np.sqrt(d))
        A_s = D_inv_sqrt @ K_PP_alpha @ D_inv_sqrt

        # Confirming A_s symmetric
        #assert np.allclose(A_s, A_s.T)
        #print(A_s - A_s.T)

        # Symmetric eigen decomposition
        eigvals, eigvecs = eigh(A_s)

        # Reverse to descending order
        eigvals = eigvals[::-1]
        eigvecs = eigvecs[:, ::-1]

        # Recover right eigenvectors of A: vi = D^{-1/2} * vi_s
        psi_pivots = D_inv_sqrt @ eigvecs

        # Store components (retain first trivial eigenvector and eigenvalue)
        self.lambdas = eigvals[0:self.n_components + 1]
        self.psi_pivots = psi_pivots[:, 0:self.n_components + 1]

        print('Done!')

        return self

    # Transform functions
    def transform(self, X_new):
        dist_new = cdist(X_new, self.pivots, 'sqeuclidean')
        K_new = np.exp(-dist_new / self.epsilon2)

        r_pivots = np.sum(K_new, axis=1)
        r_pivots_alpha = r_pivots ** self.alpha
        R_pivots_alpha_inv = np.diag(1.0 / r_pivots_alpha)
        
        K_new_alpha = R_pivots_alpha_inv @ K_new @ self.C_pivots_alpha_inv

        d = np.sum(K_new_alpha, axis=1)
        D_inv_sqrt = np.diag(1.0 / np.sqrt(d))
        A_new = D_inv_sqrt @ K_new_alpha

        # Nystr√∂m extension:
        # phi_new = A_new @ psi_pivots / lambdas
        embedding = (A_new @ self.psi_pivots) / self.lambdas
        return embedding

    def transform_streaming(self, file_path, group_path, batch_size):
        embedding = []

        with h5py.File(file_path, 'r') as f:
            group = f[group_path]
            num_records = len(group)
        
        print('Commencing Nystrom extension...')
        ctr=0
        for names, batch_array in self._stream_hdf5_batches(file_path, group_path, batch_size):
            
            #print(f"  Loaded batch of {len(names)} samples, shape = {batch_array.shape}")
                            
            dist_new = cdist(batch_array, self.pivots, 'sqeuclidean')
            K_new = np.exp(-dist_new / self.epsilon2)
    
            r_pivots = np.sum(K_new, axis=1)
            r_pivots_alpha = r_pivots ** self.alpha
            R_pivots_alpha_inv = np.diag(1.0 / r_pivots_alpha)
            
            K_new_alpha = R_pivots_alpha_inv @ K_new @ self.C_pivots_alpha_inv
    
            d = np.sum(K_new_alpha, axis=1)
            D_inv_sqrt = np.diag(1.0 / np.sqrt(d))
            A_new = D_inv_sqrt @ K_new_alpha
    
            # Nystr√∂m extension:
            # phi_new = A_new @ psi_pivots / lambdas
            embedding.append((A_new @ self.psi_pivots) / self.lambdas)

            ctr+=len(names)
            print('    %d of %d points processed' % (ctr, num_records))

        print('Done!')
        return np.concatenate(embedding)

    # Fit + transform functions
    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)

    def fit_transform_streaming(self, file_path, group_path, batch_size):
        self.fit_streaming(file_path, group_path, batch_size)
        return self.transform_streaming(file_path, group_path, batch_size)
```

### -- Swiss roll --

```python
# Generate Swiss Roll data
#X_train, color_train = make_swiss_roll(n_samples=5000, noise=0.05)

# Load Swiss Roll data from hdf5 file
file_path = './swiss_roll_data.hdf5'
group_path = 'ir'

stream_flag = True
batch_size = 512

# Fitting and embedding training data
if (stream_flag == False):
       
    with h5py.File(file_path, 'r') as f:
    
        all_keys = list(f[group_path].keys())
        print(f"Found {len(all_keys)} datasets under {group_path}")
    
        data = []
        for key in all_keys[:]:
            vec = f[f'ir/{key}'][:]
            data.append(vec)
    
        data = np.stack(data)
        print("Loaded shape:", data.shape)
    
    X_train = data
    
    # Fit Pivot Diffusion Maps with symmetric adjoint diagonalization
    pdm = PivotDiffusionMap(n_components=2, alpha=0.5, epsilon=2.)
    
    #pdm.fit(X_train)
    #embedding_train = pdm.transform(X_train)
    embedding_train = pdm.fit_transform(X_train)
    
    lambdas = pdm.lambdas
    pivots = pdm.pivots

else:

    # Fit Pivot Diffusion Maps with symmetric adjoint diagonalization
    pdm = PivotDiffusionMap(n_components=2, alpha=0.5, epsilon=2.)
   
    #pdm.fit_streaming(file_path, group_path, batch_size)
    #embedding_train = pdm.transform_streaming(file_path, group_path, batch_size)
    embedding_train = pdm.fit_transform_streaming(file_path, group_path, batch_size)
    
    lambdas = pdm.lambdas
    pivots = pdm.pivots
```

```python
# Nystrom extension on new data
if (stream_flag == False):
    
    X_new = X_train #make_swiss_roll(n_samples=300, noise=0.05)
    embedding_new = pdm.transform(X_new)

else:

    embedding_new = pdm.transform_streaming(file_path, group_path, batch_size)
```
